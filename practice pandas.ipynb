```python
# .rolling() ~ appliead to series of data by specifying window=n argument and apply the appropriate statistical function on the top of it.
a = pd.DataFrame(np.random.randn(10,4),index=pd.date_range('1/10/2022',periods=10),columns=["a1","a2","a3","a4"])
print(a)
print(a.rolling(window=5).mean())
```

```python
# .expanding() ~ it ca n be appliead on a series of data by specifying min_periods = n argument and apply  the appropriate  statistical function to the top of it
a = pd.DataFrame(np.random.randn(12,5),index=pd.date_range('1/10/2022',periods=12),columns=["a1","a2","a3","a4","a5"])
print(a)
print(a.expanding(min_periods=4).mean())
```

```python
# melt() ~ used to convert wide dataframes to narrow ones.
a = pd.DataFrame({'A':{0:'a',1:'b',2:'c'},
                  'B':{0:1,1:3,2:5},
                  'C':{0:2,1:4,2:6}})
print(a)
print(pd.melt(a,id_vars=['A'],value_vars=['B'])) 
#display only A column (as Id) with B (as variable) & corresponding value of B column (as value)
print(pd.melt(a,id_vars=['A'],value_vars=['B','C']))
 # display only A column (as Id) with B & C (as variable) & corresponding value of B and C (as value)
print(pd.melt(a,id_vars=['A'],value_vars=['B'],var_name='MyVarname',value_name='myVarName'))
#display only A column (as Id) with B (as variable) & corresponding value of B column (as value) and set variable as variablename and value as variableName
```

```python
# pivot() ~ it is a function the returns a shaped Dataframe organized organized by given index/column values
df=pd.DataFrame({'C':['one','two','three','four','five','six'],
                 'B':['A','B','C','D','E','F'],
                 'A':[1,2,3,4,5,6],
                 'D':['a','b','c','d','e','f']})
print(df.pivot(index='A',columns='B')['C'])
print(df.pivot(index='A',columns='B',values=['C','D']))
```

```python
df=pd.DataFrame({'C':['one','one','three','three','five','five'],
                 'B':['B','B','D','D','F','F'],
                 'A':[1,6,1,6,1,6],
                 'D':['a','b','c','d','e','f']})
print(df.pivot(index='A',columns='B')['C'])
print(df.pivot(index='D',columns='B',values=['C','D']))
```

```python
# Memory_usage() ~ this is a function that returns how many memory is used by each column in bytes. It is especially used when working with large data frames.
df_memory = pd.DataFrame({'One':np.random.randn(100000),
                          'Two':np.random.randint(100,size=100000)})
print(df_memory.shape)
print(df_memory)
print(df_memory.memory_usage())
print("Toal memory used in MB: ",df_memory.memory_usage().sum()/(1024**2))
```
